# 회귀 - 보스톤 주택가격 예측

```python
import numpy as np
import pandas as pd
import tensorflow as tf
seed = 2022
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, test_size=0.1, random_state=seed
)
```

-모델 정의/설정/학습

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(30, input_dim=13, activation='relu'),
    Dense(12, activation='relu'),
    Dense(1)  # 회귀의 경우 연속적인 값이기 때문에 activation 함수를 지정하지 않음
])  # 이진분류에서는 sigmoid, 다중분류에서는 softmax
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
# 회귀일 경우 metrics를 지정하지 않음
# 정확도가 없기 때문에
```

```python
hist = model.fit(X_train, y_train, validation_split=0.1, epochs=500, batch_size=60, verbose=0)
```

- 학습과정 시각화

```python
y_loss = hist.history['loss'][5:]
y_vloss = hist.history['val_loss'][5:]
xs = np.arange(6, len(y_loss)+6)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(xs, y_loss, label='train loss')
plt.plot(xs, y_vloss, label='validation loss')
plt.legend()
plt.xlabel('Epoch')
plt.grid()
plt.show()
```

- 평가 및 예측

```python
model.evaluate(X_test, y_test)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
mse = mean_squared_error(y_test, pred)
r2 = r2_score(y_test, pred)
mse, r2
```
